```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context="poster", style='ticks')

%matplotlib inline
```

# Signatures

```python
meta = pd.read_table("./preliminary/sig-DMD.union.includeERR.reference", sep="\t", header=None, usecols=[3, 5, 6])
meta.columns = ['sig-DMD', 'pair', 'pairs_number']

meta['BRCA'] = meta['pair'].apply(lambda x: 'Yes' if ('normal_trip' in x.split(",")) & 
                                  ('normal_nontrip' in x.split(",")) else "No")
meta['nonTN-BRCA'] = meta['pair'].apply(lambda x: 'Yes' if ('nontrip_trip' in x.split(",")) & 
                                        ('normal_nontrip' in x.split(",")) else "No")
meta['TN-BRCA'] = meta['pair'].apply(lambda x: 'Yes' if ('nontrip_trip' in x.split(",")) & 
                                     ('normal_trip' in x.split(",")) else "No")
meta['EndoR-BRCA'] = meta['pair'].apply(lambda x: 'Yes' if ('ers_err' in x.split(",")) else "No")

meta.head()
```

```python
methylation = pd.read_table("./datasets/methylation_sigDMD.HM450.tsv", sep="\t", usecols=['sig-DMD'])
methylation = methylation.drop_duplicates()

methylation = pd.merge(methylation, meta, how='inner', on='sig-DMD')

print(methylation.shape)
methylation.head()
```

```python
meth_expr = pd.read_table("./datasets/gene_sigDMD_corr.tsv", sep="\t")

meth_expr_tested = pd.merge(pd.DataFrame(meth_expr['sig-DMD'].unique(),columns=['sig-DMD']), meta, 
                            how='inner', on='sig-DMD')

print(meth_expr_tested.shape)
meth_expr_tested.head()
```

```python
meth_expr_corr = pd.merge(meth_expr.loc[(meth_expr['FDR_BH_p'] < 0.01) & 
                                        (meth_expr['abs_corr_coef'] >= 0.2), ['sig-DMD']].drop_duplicates(), 
                          meta, how='inner', on='sig-DMD')

print(meth_expr_corr.shape)
meth_expr_corr.head()
```

```python
group = ['BRCA', 'nonTN-BRCA', 'TN-BRCA', 'EndoR-BRCA']
total = [len(meta.loc[meta[x]=='Yes', x]) for x in group]
meth = [len(methylation.loc[methylation[x]=='Yes', x]) for x in group]
meth_tested = [len(meth_expr_tested.loc[meth_expr_tested[x]=='Yes', x]) for x in group]
meth_corr = [len(meth_expr_corr.loc[meth_expr_corr[x]=='Yes', x]) for x in group]

group.append('Union')
total.append(meta.shape[0])
meth.append(methylation.shape[0])
meth_tested.append(meth_expr_tested.shape[0])
meth_corr.append(meth_expr_corr.shape[0])

data = list(zip(group, ["Total"]*len(group), total)) + \
        list(zip(group, ["HM450k data available"]*len(group), meth)) + \
        list(zip(group, ["Tested for methylation-expression correlation"]*len(group), meth_tested)) + \
        list(zip(group, ["Methylation-expression correlated"]*len(group), meth_corr))

numbers_meth = pd.DataFrame(data, columns=['group', 'set', 'number'])
print(numbers_meth.shape)
numbers_meth
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
g = sns.barplot(data=numbers_meth, x='group', y='number', hue='set', 
                order=['Union', 'BRCA', 'nonTN-BRCA', 'TN-BRCA', 'EndoR-BRCA'],
                palette='GnBu_d')
g.set_xticklabels(['Union', 'BRCA', 'nonTN-BRCA', 'TN-BRCA', 'EndoR-BRCA'], rotation=30)
g.set_xlabel("DMV signature type")
g.set_ylabel("Number of signatures")

plt.legend(loc="best")

fig.savefig("plots/number_signatures.svg", transparent=True, bbox_inches="tight")
```

# Genes

```python
meta = pd.read_table("./metadata/associated_genes_list.tsv", sep="\t")
meta.columns = ['gene_symbol', 'BRCA', 'TN-BRCA', 'nonTN-BRCA', 'EndoR-BRCA']
print(meta.shape)
meta.head()
```

```python
expression = pd.read_table("./datasets/expression.IlluminaHiSeq.tsv", sep="\t", 
                           usecols=['gene_symbol'])
expression = expression.drop_duplicates()

expression = pd.merge(expression, meta, how='inner', on='gene_symbol')

print(expression.shape)
expression.head()
```

```python
expr_meth = pd.read_table("./datasets/gene_sigDMD_corr.tsv", sep="\t")

expr_meth_tested = pd.merge(pd.DataFrame(expr_meth['gene_symbol'].unique(),columns=['gene_symbol']), meta, 
                            how='inner', on='gene_symbol')

print(expr_meth_tested.shape)
expr_meth_tested.head()
```

```python
expr_meth_corr = pd.merge(expr_meth.loc[(expr_meth['FDR_BH_p'] < 0.01) & 
                                        (expr_meth['abs_corr_coef'] >= 0.2), ['gene_symbol']].drop_duplicates(), 
                          meta, how='inner', on='gene_symbol')

print(expr_meth_corr.shape)
expr_meth_corr.head()
```

```python
group = ['BRCA', 'nonTN-BRCA', 'TN-BRCA', 'EndoR-BRCA']
total = [len(meta.loc[meta[x]=='Yes', x]) for x in group]
expr = [len(expression.loc[expression[x]=='Yes', x]) for x in group]
expr_tested = [len(expr_meth_tested.loc[expr_meth_tested[x]=='Yes', x]) for x in group]
expr_corr = [len(expr_meth_corr.loc[expr_meth_corr[x]=='Yes', x]) for x in group]

group.append('Union')
total.append(meta.shape[0])
expr.append(expression.shape[0])
expr_tested.append(expr_meth_tested.shape[0])
expr_corr.append(expr_meth_corr.shape[0])


data = list(zip(group, ["Total"]*len(group), total)) + \
        list(zip(group, ["RNA-Seq data available"]*len(group), expr)) + \
        list(zip(group, ["Tested for methylation-expression correlation"]*len(group), expr_tested)) + \
        list(zip(group, ["Methylation-expression correlated"]*len(group), expr_corr))
numbers_expr = pd.DataFrame(data, columns=['group', 'set', 'number'])
print(numbers_expr.shape)
numbers_expr
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
g = sns.barplot(data=numbers_expr, x='group', y='number', hue='set', 
                order=['Union', 'BRCA', 'nonTN-BRCA', 'TN-BRCA', 'EndoR-BRCA'],
                palette='GnBu_d')
g.set_xticklabels(['Union', 'BRCA', 'nonTN-BRCA', 'TN-BRCA', 'EndoR-BRCA'], rotation=30)
g.set_xlabel("DMV signature type")
g.set_ylabel("Number of genes")

plt.legend(loc="best")

fig.savefig("plots/number_genes.svg", transparent=True, bbox_inches="tight")
```

# Correlated genes - signatures

```python
bigdf = pd.read_table("./datasets/correlated_gene_bigtable.tsv", sep="\t", usecols=[0,1,9,10,11,12])
bigdf.columns = ['gene_symbol', 'sig-DMD', 'BRCA', 'TN-BRCA', 'nonTN-BRCA', 'EndoR-BRCA']
print(bigdf.shape)
bigdf.head()
```

```python
from itertools import groupby

groups = ['Union', 'BRCA', 'nonTN-BRCA', 'TN-BRCA', 'EndoR-BRCA']
gene_per_sig = []
sig_per_gene = []

for gr in groups:
    if (gr != "Union"):
        sigs = list(bigdf.loc[bigdf[gr]=='Yes', 'sig-DMD'])
        genes = list(bigdf.loc[bigdf[gr]=='Yes', 'gene_symbol'])
    else:
        sigs = list(bigdf['sig-DMD'])
        genes = list(bigdf['gene_symbol'])    
    
    gps = [len(list(group)) for key, group in groupby(sigs)]
    for i in [1, 2]:
        count = len([x for x in gps if x==i])
        gene_per_sig.append((gr, str(i), count))
    
    spg = [len(list(group)) for key, group in groupby(genes)]
    for i in [1, 2, 3, 4]:
        if (i==4):
            count = len([x for x in spg if x>3])
            sig_per_gene.append((gr, ">= 4", count))
        else:
            count = len([x for x in spg if x==i])
            sig_per_gene.append((gr, str(i), count))
```

```python
gps_df = pd.DataFrame(gene_per_sig, columns=['type', 'group', 'count'])
spg_df = pd.DataFrame(sig_per_gene, columns=['type', 'group', 'count'])
```
